# Task 1: Tensor Manipulations


```python
#two sample 3D tensors

tensor1 = torch.rand(3,3,3)

tensor2 = torch.rand(3,3,3)
```

```python
print(tensor1) 
print('-'*40)
print(tensor2)
```

```python
## element wise addition
tensor_sum = tensor1 + tensor2
tensor_sum
```

```python
#element wise multiplication
tensor_product = tensor1 * tensor2
tensor_product
```

```python
# reshaping tensors
reshaped_tensor = tensor1.view(9,3)
reshaped_tensor
```

```python
# slicing and indexing tensors
print(tensor1)

'''

syntax for slicing a 3D tensor:
tensor[tensor_position_start:tensor_position_end, tensor_dimension_start:tensor_dimension_end,
         tensor_value_start:tensor_value_end]

'''

sliced_tensor = tensor1[:, 1, :] #slicing the second dimension
print(sliced_tensor) 

indexed_element = tensor1[0, 1, 0]
print(indexed_element)
```

```python
#Combining tensors

combined_tensor = torch.cat((tensor1, tensor2), dim=0) #combining along the first dimension
print(combined_tensor)
print(combined_tensor.shape)
```

```python
#splitting tensors
split_tensors = torch.chunk(combined_tensor, 2, dim=0) #splitting along the first dimension

ten1 , ten2 = split_tensors
print(ten1)
print(ten2)
```

# TASK 2: Neural network using Keras

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Flatten(input_shape=(28, 28)))
model.add(tf.keras.layers.Dense(128, activation='relu'))
model.add(tf.keras.layers.Dense(10, activation='softmax'))
```

```python
#compiling the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

# TASK 3: Classification Tasks:
## Build and train a fully connected neural network for classification tasks using first MNIST and then IMDb datasets

```python
from tensorflow.keras.datasets import mnist
```

```python
(x_train_full, y_train_full), (x_test, y_test) = mnist.load_data()
x_train = x_train_full[:50000]
y_train = y_train_full[:50000]
x_valid = x_train_full[50000:]
y_valid = y_train_full[50000:]
(x_train.shape, y_train.shape), (x_valid.shape, y_valid.shape), (x_test.shape, y_test.shape)
```

Normalize

```python
x_train, x_valid, x_test = x_train / 255.0, x_valid / 255.0, x_test / 255.0
```

```python
matplotlib.pyplot.imshow(x_train[0], cmap='gray')
matplotlib.pyplot.axis('off')
matplotlib.pyplot.show()
```

```python
y_train[0]
```

```python
mnist_classifier = tf.keras.models.clone_model(model)

mnist_classifier.summary()
```

```python
mnist_model_history = mnist_classifier.fit(x_train, y_train, epochs=5, validation_data=(x_valid, y_valid))
```

 Plot accuracy

```python
matplotlib.pyplot.figure(figsize=(15,5))
matplotlib.pyplot.subplot(1,2,1)
matplotlib.pyplot.plot(mnist_model_history.history['accuracy'], label='Train Accuracy')
matplotlib.pyplot.plot(mnist_model_history.history['val_accuracy'], label='Validation Accuracy')

matplotlib.pyplot.xlabel('Epoch')
matplotlib.pyplot.ylabel('Accuracy')
matplotlib.pyplot.title('Training vs Validation Accuracy')
matplotlib.pyplot.legend()
matplotlib.pyplot.grid(True)
matplotlib.pyplot.subplot(1,2,2)
matplotlib.pyplot.plot(mnist_model_history.history['loss'], label='Train Loss')
matplotlib.pyplot.plot(mnist_model_history.history['val_loss'], label='Validation Loss')

matplotlib.pyplot.xlabel('Epoch')
matplotlib.pyplot.ylabel('Loss')
matplotlib.pyplot.title('Training vs Validation Loss')
matplotlib.pyplot.legend()
matplotlib.pyplot.grid(True)
matplotlib.pyplot.show()
```

plot loss

```python
matplotlib.pyplot.figure(figsize=(8,5))
matplotlib.pyplot.plot(mnist_model_history.history['loss'], label='Train Loss')
matplotlib.pyplot.plot(mnist_model_history.history['val_loss'], label='Validation Loss')

matplotlib.pyplot.xlabel('Epoch')
matplotlib.pyplot.ylabel('Loss')
matplotlib.pyplot.title('Training vs Validation Loss')
matplotlib.pyplot.legend()
matplotlib.pyplot.grid(True)
matplotlib.pyplot.show()
```

Evaluate the model and display test loss and accuracy

```python
test_loss, test_acc = mnist_classifier.evaluate(x_test,  y_test, verbose=2)

print(f'\nTest accuracy: {test_acc*100: .4f} %')
print(f'Test loss: {test_loss*100: .4f} %')
```

Data preperation

```python
# padding, cause not all reviews have same length
for i, review in enumerate(x_train[1000:]):
    print(len(review),',', end='')
    if i == 10:
        break

#also they are the same size
print('\n',x_train.shape)
print(x_test.shape)
```

```python

# x_test dont need that much data, so lets give it to x_train

x_full = np.concatenate((x_train, x_test), axis=0)
y_full = np.concatenate((y_train, y_test), axis=0)

print(x_full.shape)
print(y_full.shape)


```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_full, y_full, test_size=0.1, random_state=42)
x_train, x_valid, y_train, y_valid = train_test_split(x_train, y_train, test_size=0.1, random_state=42)

print('\n',x_train.shape)
print(x_valid.shape)   
print(x_test.shape)

```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
x_train = pad_sequences(x_train, maxlen=300)
x_valid = pad_sequences(x_valid, maxlen=300)
x_test = pad_sequences(x_test, maxlen=300)

for i, review in enumerate(x_train[1000:]):
    print(len(review),',', end='')
    if i == 10:
        break
```

```python
x_train.shape
```

lets build simple model now

```python

imdb_model = tf.keras.models.Sequential()
imdb_model.add(tf.keras.layers.Dense(300,input_shape=(300,)))
imdb_model.add(tf.keras.layers.Dense(128, activation='relu'))
imdb_model.add(tf.keras.layers.Dense(64, activation='relu'))
imdb_model.add(tf.keras.layers.Dense(2, activation='softmax'))

imdb_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
imdb_model.summary()
```

```python
imdb_model_history = imdb_model.fit(x_train, y_train, epochs=5,validation_data=(x_valid, y_valid))
```

```python
#plot training vs validation accuracy and loss

matplotlib.pyplot.figure(figsize=(15,5))
matplotlib.pyplot.subplot(1,2,1)
matplotlib.pyplot.plot(imdb_model_history.history['accuracy'], label='Train Accuracy')
matplotlib.pyplot.plot(imdb_model_history.history['val_accuracy'], label='Validation Accuracy')

matplotlib.pyplot.xlabel('Epoch')
matplotlib.pyplot.ylabel('Accuracy')
matplotlib.pyplot.title('Training vs Validation Accuracy')
matplotlib.pyplot.legend()
matplotlib.pyplot.grid(True)

matplotlib.pyplot.subplot(1,2,2)
matplotlib.pyplot.plot(mnist_model_history.history['loss'], label='Train Loss')
matplotlib.pyplot.plot(mnist_model_history.history['val_loss'], label='Validation Loss')

matplotlib.pyplot.xlabel('Epoch')
matplotlib.pyplot.ylabel('Loss')
matplotlib.pyplot.title('Training vs Validation Loss')
matplotlib.pyplot.legend()
matplotlib.pyplot.grid(True)
matplotlib.pyplot.show()
matplotlib.pyplot.show()
```
